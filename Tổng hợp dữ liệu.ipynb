```python
last_trans_account = completed_trans_df.drop_duplicates(subset='account_id', keep='last')
```

```python
first_trans_account = completed_trans_df.drop_duplicates(subset='account_id', keep='first')
```

```python
trans_account_synthesis = pd.merge(first_trans_account, last_trans_account, on='account_id', how='inner', suffixes= ('_first', '_last'))
```

```python
trans_account_synthesis['fulldate_last_pre'] = pd.to_datetime(trans_account_synthesis['fulldate_last'], format='%Y-%m-%d')
trans_account_synthesis['fulldate_first_pre'] = pd.to_datetime(trans_account_synthesis['fulldate_first'], format='%Y-%m-%d')
trans_account_synthesis['diff_date'] = (trans_account_synthesis['fulldate_last_pre'] - trans_account_synthesis['fulldate_first_pre']).dt.days
```

```python
columns_used = ['account_id', 'balance_last', 'diff_date','date_last']
trans_last_synthesis_df = trans_account_synthesis[columns_used]
```
